```python
df=pandas.read_csv("/Users/fowler/packages/tuberculosis_amr_catalogues/catalogues/NC_000962.3/NC_000962.3_CRyPTIC_v1.2_GARC1_RUS.csv")
```

```python
df
```

```python
def split_mutation(row):
    
    components=row['MUTATION'].split("@")

    # the gene name is always the first component
    gene=components[0]

    # ..and the remainder is the mutation
    mutation=row['MUTATION'].split(gene+"@")[1]
    
    return(pandas.Series([gene,mutation]))

df[['GENE','MUTATION']]=df.apply(split_mutation,axis=1)
```

```python
cats=df.loc[df.PREDICTION=="R"][['DRUG','GENE']].groupby(['DRUG','GENE']).count()
cats
```

```python
df.set_index(['DRUG','GENE'],inplace=True)
df
```

```python
df.join(cats,how='right')
```
